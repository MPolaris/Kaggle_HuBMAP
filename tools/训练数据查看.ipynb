```python
import os
import cv2
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
target_root = "E:/HuBMAP_data/"
img_root = target_root + "img/"
mask_root = target_root + "mask/"
filelist = os.listdir(img_root)
```

```python
fn = np.random.choice(filelist)
image = cv2.imread(img_root + fn)
maskname = mask_root + fn.split("_")[0] + "_mask.png"
if os.path.exists(maskname):
    print(maskname)
    mask = cv2.imread(mask_root + fn.split("_")[0] + "_mask.png")
    mask = mask[:,:,0]/255
else:
    mask = np.zeros_like(image)

plt.figure(figsize=(30,30))
plt.subplot("121")
plt.axis('off')
plt.imshow(image)
plt.imshow(mask,alpha=0.5, cmap='gray')
plt.subplot("122")
plt.axis('off')
plt.imshow(image)
plt.show()
```
